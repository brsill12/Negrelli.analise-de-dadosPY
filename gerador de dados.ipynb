Analise de dados de cartao de credito

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

Ler a base e organiza-la


```python
dados = pd.read_csv('fatura.csv')
```

Iniciar o processo de analise

valor total da fatura 

```python
valor_total_cartao = dados['ValorCompra'].sum()
print(f"O valor total da fatura é: R$ {valor_total_cartao}")
```

valor gasto por titutal/pessoa

```python
dados
```

```python
valor_gasto_por_titular = dados.groupby('Nome')['ValorCompra'].sum().reset_index().sort_values(by = 'ValorCompra', ascending=False)
```

Valor gasto por categoria

```python
valor_gasto_por_categoria = dados.groupby('categoria')['ValorCompra'].sum().reset_index().sort_values(by = 'ValorCompra', ascending=False) 
valor_gasto_por_categoria
```

Dias que mais ocorreram gastos

```python
gastos_por_dia = dados.groupby('data_compra')['ValorCompra'].sum().reset_index().sort_values(by = 'data_compra', ascending=False) 
gastos_por_dia
```

Grafico Gastos por pessoa

```python
plt.figure(figsize=(10,6))
plt.bar(valor_gasto_por_titular['Nome'], valor_gasto_por_titular['ValorCompra'], color='black')
plt.title("valor gasto por pessoa")
plt.xlabel('titular')
plt.ylabel('Valor gasto (R$)')
plt.xticks(rotation=45)
plt.show()         
```

Grafico de gastos por categoria 

```python
plt.figure(figsize=(10,6))
plt.barh(valor_gasto_por_categoria['categoria'], valor_gasto_por_categoria['ValorCompra'])
plt.show()
```

Grafico por gastos de titular

```python
plt.figure(figsize=(5,5))
plt.title("Gastos por titular")
plt.pie(valor_gasto_por_titular['ValorCompra'],labels=valor_gasto_por_titular['Nome'], autopct = '%1.1f%%')
plt.show()
```
